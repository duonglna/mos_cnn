```python
import os
import sys
import numpy as np
import pandas as pd
```

```python

res_dir='/home/adam/VecTech/repos/novel-species-detection/tierIII_output/run75/'
odin_res_dir=res_dir+'odin2'
open_res_dir=res_dir+'openset_breakdown'
fold_to_file_to_set = {}

for root,dirs,files in os.walk(open_res_dir):
    # print(root, files)
    for f in files:
        if f.split('.')[-1]=='csv':
            fold_to_file_to_set[f]={}
            tmp_df=pd.read_csv(root+'/'+f)
            for i in range(len(tmp_df)):
                fold_to_file_to_set[f][tmp_df.iloc[i]['Id']]=tmp_df.iloc[i]['split']
```

```python
new_save_dir = '/home/adam/VecTech/repos/novel-species-detection/tierIII_output/run75/new_odin/'
for root,dirs,files in os.walk(odin_res_dir):
    # print(root, files)
    for f in files:
        if f.split('.')[-1]=='csv':
            tmp_df=pd.read_csv(root+'/'+f)
            f_key=f.replace('result_','openset')
            splitkey=[]
            for i in range(len(tmp_df)):
                tmp_id=tmp_df.iloc[i]['Id']
                print(f_key)
                splitkey.append(fold_to_file_to_set[f_key][tmp_id])
            tmp_df['split']=splitkey
            tmp_df.to_csv(new_save_dir+f.replace('result_','odin'))
```
